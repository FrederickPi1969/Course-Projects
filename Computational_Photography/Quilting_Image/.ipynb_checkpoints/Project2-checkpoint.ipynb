```python
def quilt_random(sample, out_size, patch_size):
    """
    Randomly samples square patches of size patchsize from sample in order to create an output image of size outsize.

    :param sample: numpy.ndarray   The image you read from sample directory
    :param out_size: int            The width of the square output image
    :param patch_size: int          The width of the square sample patch
    :return: numpy.ndarray
    """
    
    # Todo 
    output = np.zeros([out_size, out_size, 3])
    for i in range(0, out_size, patch_size):
        if out_size - i < patch_size:
            break
        for j in range(0, out_size, patch_size):
            if out_size - i < patch_size or out_size - j < patch_size:
                continue
            rand_x = np.random.randint(0, sample.shape[0] - patch_size)
            rand_y = np.random.randint(0, sample.shape[0] - patch_size)
            patch = sample[rand_x:rand_x + patch_size, rand_y : rand_y + patch_size, :]
            output[i:i + patch_size, j : j + patch_size,:] = patch
    return output
```

### Part II: Overlapping Patches (30 pts)

```python
def ssd_patch(M, T, I):
    ## ssd 
    ssd = ((M*T)**2).sum() - 2 * cv2.filter2D(I, ddepth=-1, kernel = M*T) + cv2.filter2D(I ** 2, ddepth=-1, kernel=M)
    return ssd
    
def choose_sample(sample, patch_size, M, T, true_sample, tol):
    cost_image = np.zeros([sample.shape[0] - patch_size, sample.shape[0]- patch_size])
    for i in range(sample.shape[0] - patch_size):
        for j in range(sample.shape[0] - patch_size):
            cost_image[i,j] = np.sum(ssd_patch(M, T, sample[i:i+patch_size, j : j + patch_size]))
#     tenth_largest = np.sort(cost_image.ravel())[-10]
#     x,y = np.where(cost_image >= tenth_largest)
#     rand = np.random.randint(0,10)
#     x,y = (x[rand], y[rand])
    min_value = np.min(cost_image)
    row, col = np.where(cost_image <= min_value *(1+tol))
    index = np.random.randint(0, len(row))
    x,y = row[index], row[index]
    return sample[x : x + patch_size, y : y + patch_size], true_sample[x : x + patch_size, y : y + patch_size, :]
```

```python
def put_new_sample(i, j, overlap, patch_size, new_sample, sample):
    if i == 0 and j == 0:
        sample[i : i + patch_size, j : j + patch_size] = new_sample
    elif j == 0 and i != 0:
        sample[i + overlap // 2 : i + patch_size, j : j + patch_size] = new_sample[overlap // 2:, :, :]
    elif j != 0 and i == 0:
        
        sample[i: i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[:, overlap // 2:, :]
    else:
        sample[i + overlap // 2 : i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[overlap // 2:, overlap // 2:, :]

def put_grey_sample(i, j, overlap, patch_size, new_sample, sample):
    if i == 0 and j == 0:
        sample[i : i + patch_size, j : j + patch_size] = new_sample
    elif j == 0 and i != 0:
        sample[i + overlap // 2 : i + patch_size, j : j + patch_size] = new_sample[overlap // 2:, :]
    elif j != 0 and i == 0:
    
        sample[i: i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[:, overlap // 2:]
    else:
        sample[i + overlap // 2 : i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[overlap // 2:, overlap // 2:]
```

```python
def quilt_simple(sample, out_size, patch_size, overlap, tol):
    """
    Randomly samples square patches of size patchsize from sample in order to create an output image of size outsize.
    Feel free to add function parameters
    :param sample: numpy.ndarray
    :param out_size: int
    :param patch_size: int
    :param overlap: int
    :param tol: int
    :return: numpy.ndarray
    """
    assert overlap < patch_size
    # Todo 
    grey = cv2.cvtColor(sample, cv2.COLOR_RGB2GRAY) / 255.0
    step = patch_size - overlap
    grey_output = np.zeros([out_size, out_size])
    output = np.zeros([out_size, out_size, 3])
    
    for i in range(0, out_size + 1, step):
        if out_size - i < patch_size:
            break
        for j in range(0, out_size + 1, step):
            print(i, j)
            if out_size - j < patch_size:
                continue
            M = (grey_output[i : i + patch_size, j : j + patch_size] != 0).astype(int)
            T = (grey_output[i : i + patch_size, j : j + patch_size])
            grey_sample, new_sample = choose_sample(grey, patch_size, M, T, sample, tol)
            put_grey_sample(i, j, overlap, patch_size, grey_sample, grey_output)
            put_new_sample(i, j, overlap, patch_size, new_sample, output)
    return output
            
```

```python
def get_all_patches(sample, patch_size):
    all_patches = {}
    count = -1
    for i in range(sample.shape[0]):
        if sample.shape[0] - i < patch_size:
            break
        for j in range(sample.shape[1]):
            if sample.shape[1] - j < patch_size:
                continue
            count += 1
            all_patches[count] = sample[i : i + patch_size, j : j + patch_size, :]
    return all_patches
            
        
def ssd_cost(M, T, I):
    useful_I = np.zeros(I.shape)
    useful_I[:,:,0], useful_I[:,:,1], useful_I[:,:,2] = M * I[:,:,0], M * I[:,:,1], M * I[:,:,2]
    return np.sum((T - useful_I) ** 2)

    
def choose_sample(sample, patch_size, M, T, tol, all_patches):
#     cost_image = np.zeros([sample.shape[0] - patch_size, sample.shape[0]- patch_size])
    proportion = 0.2
    cost_image = {}
    n = len(all_patches)
    index = np.random.choice(n, int(proportion * n))

    for i in index:
        cost_image[i] = ssd_cost(M, T, all_patches[i])
    
    min_value = min(cost_image.values())
    choices = []
    for key in cost_image.keys():
        if cost_image[key] <= min_value * (1 + tol):
            choices.append(all_patches[key])
    
    return choices[np.random.choice(len(choices))]
```

```python
def get_horizontal_mask(cost, overlap):
    template = cost[:, 0 : overlap]
    return cut(template.T).T
    
def get_vertical_mask(cost, overlap):
    template = cost[0 : overlap, :]
    return cut(template)

def ssd_patch(M, T, I):
    ## ssd 
    ssd = ((M*T)**2).sum() - 2 * cv2.filter2D(I, ddepth=-1, kernel = M*T) + cv2.filter2D(I ** 2, ddepth=-1, kernel=M)
    return ssd
```

```python
def choose_sample_p3(sample, patch_size, M, T, true_sample, tol):
    cost_image = np.zeros([sample.shape[0] - patch_size, sample.shape[0]- patch_size])
    cost = np.zeros([sample.shape[0] - patch_size, sample.shape[1] - patch_size, patch_size, patch_size])
    for i in range(sample.shape[0] - patch_size):
        for j in range(sample.shape[0] - patch_size):
            temp = ssd_patch(M, T, sample[i:i + patch_size, j : j + patch_size])
            cost_image[i,j] = np.sum(temp)
            cost[i, j, :,:] = temp
            
    min_value = np.min(cost_image)
    row, col = np.where(cost_image <= min_value *(1+tol))
    index = np.random.randint(0, len(row))
    x,y = row[index], col[index]
    
    return sample[x : x + patch_size, y : y + patch_size], true_sample[x : x + patch_size, y : y + patch_size, :],\
           cost[x, y]

def fill_horizontal(i, j, patch_size, overlap, mask, grey_output, rgb_output, grey_sample, rgb_sample):
    overlap_grey = grey_output[i : i + patch_size, j : j + overlap]
    overlap_rgb = rgb_output[i : i + patch_size, j : j + overlap, :]
    
    overlap_grey[:,:] = mask * grey_sample[:,:overlap]  + (1 - mask) * overlap_grey
    for k in range(3):
        overlap_rgb[:,:,k] =  rgb_sample[:,:overlap,k] * mask  + (1 - mask) * overlap_rgb[:,:,k]
        
    
    
def fill_vertical(i, j, patch_size, overlap, mask, grey_output, rgb_output, grey_sample, rgb_sample):
    overlap_grey = grey_output[i : i + overlap, j : j + patch_size]
    overlap_rgb = rgb_output[i : i + overlap, j : j + patch_size, :]
    
    overlap_grey[:,:] = grey_sample[:overlap, :] * mask + (1 - mask) * overlap_grey
    for k in range(3):
        overlap_rgb[:,:,k] =  rgb_sample[:overlap,:,k] * mask  + (1 - mask) * overlap_rgb[:,:,k]
#     overlap_rgb = (rgb_sample[:overlap, :, :].T * rgb_mask).T + ((1 - rgb_mask) * overlap_rgb.T).T


def fill_leftover_horizontal(i, j, patch_size, overlap, grey_output, rgb_output, grey_sample, rgb_sample):
    grey_output[i : i + patch_size, j + overlap :j + patch_size] = grey_sample[:, overlap:]
    rgb_output[i : i + patch_size, j + overlap : j + patch_size,:] = rgb_sample[:, overlap:,:]
    
def fill_leftover_vertical(i, j, patch_size, overlap, grey_output, rgb_output, grey_sample, rgb_sample):
    grey_output[i + overlap : i + patch_size, j : j + patch_size] = grey_sample[overlap:, :]
    rgb_output[i + overlap : i + patch_size, j : j + patch_size,:] = rgb_sample[overlap:, :,:]

def fill_leftover(i, j, patch_size, overlap, grey_output, rgb_output, grey_sample, rgb_sample):
    grey_output[i + overlap : i + patch_size, j + overlap :j + patch_size] = grey_sample[overlap:, overlap:]
    rgb_output[i + overlap : i + patch_size, j + overlap : j + patch_size,:] = rgb_sample[overlap:, overlap:,:]

def put_new_sample(i, j, overlap, patch_size, new_sample, sample):
    if i == 0 and j == 0:
        sample[i : i + patch_size, j : j + patch_size] = new_sample
    elif j == 0 and i != 0:
        sample[i + overlap // 2 : i + patch_size, j : j + patch_size] = new_sample[overlap // 2:, :, :]
    elif j != 0 and i == 0:
        
        sample[i: i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[:, overlap // 2:, :]
    else:
        sample[i + overlap // 2 : i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[overlap // 2:, overlap // 2:, :]

def put_grey_sample(i, j, overlap, patch_size, new_sample, sample):
    if i == 0 and j == 0:
        sample[i : i + patch_size, j : j + patch_size] = new_sample
    elif j == 0 and i != 0:
        sample[i + overlap // 2 : i + patch_size, j : j + patch_size] = new_sample[overlap // 2:, :]
    elif j != 0 and i == 0:
    
        sample[i: i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[:, overlap // 2:]
    else:
        sample[i + overlap // 2 : i + patch_size, j + overlap // 2 : j + patch_size] = new_sample[overlap // 2:, overlap // 2:]
```

```python
def quilt_cut(sample, out_size, patch_size, overlap, tol):
    """
    Samples square patches of size patchsize from sample using seam finding in order to create an output image of size outsize.
    Feel free to add function parameters
    :param sample: numpy.ndarray
    :param out_size: int
    :param patch_size: int
    :param overlap: int
    :param tol: int
    :return: numpy.ndarray
    """
    grey = cv2.cvtColor(sample, cv2.COLOR_RGB2GRAY) / 255.0
    step = patch_size - overlap
    grey_output = np.zeros([out_size, out_size])
    output = np.zeros([out_size, out_size, 3])
    
    
    for i in range(0, out_size + 1, step):
        if out_size - i < patch_size:
            break
        for j in range(0, out_size + 1, step):
            print(i, j)
            if out_size - j < patch_size:
                continue
            M = (grey_output[i : i + patch_size, j : j + patch_size] != 0).astype(int)
            T = (grey_output[i : i + patch_size, j : j + patch_size])
            grey_sample, new_sample, cost = choose_sample_p3(grey, patch_size, M, T, sample, tol)
            if i == 0 and j == 0:
                put_grey_sample(i, j, overlap, patch_size, grey_sample, grey_output)
                put_new_sample(i, j, overlap, patch_size, new_sample, output)
                
            elif i == 0 and j != 0:
                mask = get_horizontal_mask(cost, overlap)
                fill_horizontal(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, new_sample)
                fill_leftover_horizontal(i, j, patch_size, overlap, grey_output, output, grey_sample, new_sample)
                
            elif i != 0 and j == 0:
                mask = get_vertical_mask(cost, overlap)
                fill_vertical(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, new_sample)
                fill_leftover_vertical(i, j, patch_size, overlap, grey_output, output, grey_sample, new_sample)
                
            else:
                mask = get_horizontal_mask(cost, overlap)
                fill_horizontal(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, new_sample)  
                cost = ssd_patch(M, T, grey_sample)
                mask = get_vertical_mask(cost, overlap)
                fill_vertical(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, new_sample)
                fill_leftover(i, j, patch_size, overlap, grey_output, output, grey_sample, new_sample)
                
                
    return output
```

```python
def choose_sample_p4(grey_texture, overlap, patch_size, M, T, rgb_texture, tol, grey_target):
    ## sample is texture
    cost_image = np.zeros([grey_texture.shape[0] - patch_size, grey_texture.shape[1]- patch_size])
    cost = np.zeros([grey_texture.shape[0] - patch_size, grey_texture.shape[1] - patch_size, patch_size, patch_size])
    for i in range(grey_texture.shape[0] - patch_size):
        for j in range(grey_texture.shape[1] - patch_size):
            alpha = np.sum(M) / (M.shape[0] * M.shape[1])
            cost1 = ssd_patch(M, T, grey_texture[i:i + patch_size, j : j + patch_size])
            cost2 = ssd_patch(np.ones([patch_size, patch_size]),\
                                            grey_target, grey_texture[i : i + patch_size, j : j + patch_size])
#             print("gray_target shape : ", grey_target.shape)
#             print("grey_texture shape :", grey_texture[i : i + patch_size , j + j ])
#             print("cost1 shape : ", cost1.shape, " cost 2 shape : ", cost2.shape) 
            cost_image[i,j] = alpha * np.sum(cost1) + (1 - alpha) * np.sum(cost2)
            cost[i, j, :,:] = cost1
    
    min_value = np.min(cost_image)
    row, col = np.where(cost_image <= min_value *(1+tol))
    index = np.random.randint(0, len(row))
    x,y = row[index], col[index]
    
    return grey_texture[x : x + patch_size, y : y + patch_size], rgb_texture[x : x + patch_size, y : y + patch_size, :],\
           cost[x, y]
```

```python
def texture_transfer(texture, target, patch_size, overlap, tol):
    """
    Feel free to add function parameters
    """
    grey_texture = cv2.cvtColor(texture, cv2.COLOR_RGB2GRAY) / 255.0
    step = patch_size - overlap
    grey_output = np.zeros([target.shape[0], target.shape[1]])
    output = target.copy()
    grey_target = cv2.cvtColor(target, cv2.COLOR_RGB2GRAY) / 255.0
    
    
    for i in range(0, target.shape[0] + 1, step):
        if target.shape[0] - i < patch_size:
            break
        for j in range(0, target.shape[1] + 1, step):
            print(i, j)
            if target.shape[1] - j < patch_size:
                continue
            M = (grey_output[i : i + patch_size, j : j + patch_size] != 0).astype(int)
            T = (grey_output[i : i + patch_size, j : j + patch_size])
            grey_sample, rgb_texture_sample, cost = choose_sample_p4(grey_texture, overlap, patch_size, M, T,\
                                                             texture, tol, grey_target[i:i + patch_size, j : j + patch_size])
            if i == 0 and j == 0:
                print(grey_sample.shape)
                put_grey_sample(i, j, overlap, patch_size, grey_sample, grey_output)
                put_new_sample(i, j, overlap, patch_size, rgb_texture_sample, output)
                
            elif i == 0 and j != 0:
                mask = get_horizontal_mask(cost, overlap)
                fill_horizontal(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, rgb_texture_sample)
                fill_leftover_horizontal(i, j, patch_size, overlap, grey_output, output, grey_sample,  rgb_texture_sample)
                
            elif i != 0 and j == 0:
                mask = get_vertical_mask(cost, overlap)
                fill_vertical(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, rgb_texture_sample)
                fill_leftover_vertical(i, j, patch_size, overlap, grey_output, output, grey_sample,  rgb_texture_sample)
                
            else:
                mask = get_horizontal_mask(cost, overlap)
                fill_horizontal(i, j, patch_size, overlap, mask, grey_output, output, grey_sample, rgb_texture_sample)  
                cost = ssd_patch(M, T, grey_sample)
                mask = get_vertical_mask(cost, overlap)
                fill_vertical(i, j, patch_size, overlap, mask, grey_output, output, grey_sample,  rgb_texture_sample)
                fill_leftover(i, j, patch_size, overlap, grey_output, output, grey_sample,  rgb_texture_sample)
                
                
    return output
```
